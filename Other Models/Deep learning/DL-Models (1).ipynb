```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, LSTM, Dense, Dropout, Bidirectional, GlobalMaxPooling1D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("Deep Learning libraries imported successfully!")

# Cell 2: Load and Verify Data
print("Loading preprocessed data...")

# Load sequences (text data)
X_seq_train = np.load('X_seq_train.npy')
X_seq_test = np.load('X_seq_test.npy')
y_train = np.load('y_train.npy')
y_test = np.load('y_test.npy')

print("Data loaded successfully!")
print(f"Sequence data - Train: {X_seq_train.shape}, Test: {X_seq_test.shape}")
print(f"Labels - Train: {y_train.shape}, Test: {y_test.shape}")
print(f"Class distribution - Train: {np.unique(y_train, return_counts=True)}")
print(f"Class distribution - Test: {np.unique(y_test, return_counts=True)}")

# Model Parameters
VOCAB_SIZE = 8825
MAX_LEN = 100
EMBEDDING_DIM = 100
BATCH_SIZE = 32
EPOCHS = 30
LEARNING_RATE = 0.001

# Calculate Class Weights
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}
print(f"Class weights: {class_weight_dict}")

# Cell 5: Create Validation Set
X_seq_temp, X_seq_val, y_temp, y_val = train_test_split(
    X_seq_train, y_train, 
    test_size=0.2, 
    random_state=42, 
    stratify=y_train
)

print(f"Training: {X_seq_temp.shape}, Validation: {X_seq_val.shape}")

# Model 1 - CNN
def create_cnn_model():
    """CNN model for text classification"""
    model = Sequential([
        # Embedding layer
        Embedding(VOCAB_SIZE, EMBEDDING_DIM, input_length=MAX_LEN),
        
        # First CNN block
        Conv1D(128, 5, activation='relu', padding='same'),
        MaxPooling1D(2),
        Dropout(0.3),
        
        # Second CNN block
        Conv1D(64, 3, activation='relu', padding='same'),
        MaxPooling1D(2),
        Dropout(0.3),
        
        # Third CNN block
        Conv1D(32, 3, activation='relu', padding='same'),
        GlobalMaxPooling1D(),
        Dropout(0.3),
        
        # Dense layers
        Dense(64, activation='relu'),
        Dropout(0.4),
        Dense(32, activation='relu'),
        Dropout(0.4),
        Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=LEARNING_RATE),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Model 2 - LSTM
def create_lstm_model():
    """Standard LSTM model"""
    model = Sequential([
        # Text sequence input
        Embedding(VOCAB_SIZE, EMBEDDING_DIM, input_length=MAX_LEN, mask_zero=True),
        
        # LSTM layers
        LSTM(128, return_sequences=True, dropout=0.3, recurrent_dropout=0.2),
        LSTM(64, dropout=0.3, recurrent_dropout=0.2),
        Dropout(0.4),
        
        # Dense layers
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=LEARNING_RATE),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Model 3 - Bidirectional LSTM
def create_bilstm_model():
    """Bidirectional LSTM model"""
    model = Sequential([
        # Text sequence input
        Embedding(VOCAB_SIZE, EMBEDDING_DIM, input_length=MAX_LEN, mask_zero=True),
        
        # Bidirectional LSTM layers
        Bidirectional(LSTM(128, return_sequences=True, dropout=0.3, recurrent_dropout=0.2)),
        Bidirectional(LSTM(64, dropout=0.3, recurrent_dropout=0.2)),
        Dropout(0.4),
        
        # Dense layers
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=LEARNING_RATE),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Model 4 - CNN-LSTM Hybrid
def create_cnn_lstm_model():
    """CNN-LSTM Hybrid model"""
    model = Sequential([
        # Embedding layer
        Embedding(VOCAB_SIZE, EMBEDDING_DIM, input_length=MAX_LEN),
        
        # CNN layers for feature extraction
        Conv1D(128, 5, activation='relu', padding='same'),
        MaxPooling1D(2),
        Dropout(0.3),
        
        Conv1D(64, 3, activation='relu', padding='same'),
        MaxPooling1D(2),
        Dropout(0.3),
        
        # LSTM layers for sequence processing
        LSTM(64, return_sequences=True, dropout=0.3, recurrent_dropout=0.2),
        LSTM(32, dropout=0.3, recurrent_dropout=0.2),
        Dropout(0.4),
        
        # Dense layers
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=LEARNING_RATE),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Create and Display Models
print("Creating models...")
models = {
    'CNN': create_cnn_model(),
    'LSTM': create_lstm_model(),
    'Bi-LSTM': create_bilstm_model(),
    'CNN-LSTM': create_cnn_lstm_model()
}

for name, model in models.items():
    print(f"\n{name} Model:")
    model.summary()

# Training Function
def train_and_evaluate_model(model, model_name, X_train_seq, y_train, X_val_seq, y_val, X_test_seq, y_test):
    print(f"\n{'='*50}")
    print(f" TRAINING {model_name}")
    print(f"{'='*50}")
    
    # Callbacks
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=7,
            restore_best_weights=True,
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=4,
            min_lr=0.0001,
            verbose=1
        )
    ]
    
    # Train model
    history = model.fit(
        X_train_seq, y_train,
        validation_data=(X_val_seq, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        class_weight=class_weight_dict,
        callbacks=callbacks,
        verbose=1
    )
    
    # Make predictions on TEST set
    y_pred_proba = model.predict(X_test_seq).flatten()
    y_pred = (y_pred_proba > 0.5).astype(int)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    auc_roc = roc_auc_score(y_test, y_pred_proba)
    
    print(f"\n {model_name} RESULTS:")
    print(f"Accuracy:  {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-Score:  {f1:.4f}")
    print(f"AUC-ROC:   {auc_roc:.4f}")
    
    # Check prediction distribution
    unique_preds = np.unique(y_pred, return_counts=True)
    print(f"Predictions distribution: {dict(zip(unique_preds[0], unique_preds[1]))}")
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Ham', 'Smish'], 
                yticklabels=['Ham', 'Smish'])
    plt.title(f'{model_name} - Confusion Matrix', fontweight='bold')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    
    # Classification report
    print(f"\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=['Ham', 'Smish']))
    
    return {
        'accuracy': accuracy, 'precision': precision, 'recall': recall,
        'f1_score': f1, 'auc_roc': auc_roc, 
        'y_pred': y_pred, 'y_pred_proba': y_pred_proba,
        'history': history
    }

# Train All Models
print("Starting model training...")
results = {}

for name, model in models.items():
    results[name] = train_and_evaluate_model(
        model, name, 
        X_seq_temp, y_temp, 
        X_seq_val, y_val, 
        X_seq_test, y_test
    )

# Comprehensive Model Comparison
print("\n" + "="*70)
print("DEEP LEARNING MODEL COMPARISON")
print("="*70)

comparison_df = pd.DataFrame({
    model: {
        'Accuracy': metrics['accuracy'],
        'Precision': metrics['precision'],
        'Recall': metrics['recall'],
        'F1-Score': metrics['f1_score'],
        'AUC-ROC': metrics['auc_roc']
    }
    for model, metrics in results.items()
}).T

print(comparison_df.round(4))

# Cell 14: Performance Visualization
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Accuracy comparison
models_names = list(comparison_df.index)
accuracy_scores = [comparison_df.loc[model, 'Accuracy'] for model in models_names]
bars1 = axes[0,0].bar(models_names, accuracy_scores, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
axes[0,0].set_title('Model Accuracy Comparison', fontsize=14, fontweight='bold')
axes[0,0].set_ylabel('Accuracy')
axes[0,0].set_ylim(0, 1)
for bar in bars1:
    height = bar.get_height()
    axes[0,0].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                  f'{height:.3f}', ha='center', va='bottom')

# F1-Score comparison
f1_scores = [comparison_df.loc[model, 'F1-Score'] for model in models_names]
bars2 = axes[0,1].bar(models_names, f1_scores, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
axes[0,1].set_title('Model F1-Score Comparison', fontsize=14, fontweight='bold')
axes[0,1].set_ylabel('F1-Score')
axes[0,1].set_ylim(0, 1)
for bar in bars2:
    height = bar.get_height()
    axes[0,1].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                  f'{height:.3f}', ha='center', va='bottom')

# Recall comparison
recall_scores = [comparison_df.loc[model, 'Recall'] for model in models_names]
bars3 = axes[0,2].bar(models_names, recall_scores, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
axes[0,2].set_title('Model Recall Comparison', fontsize=14, fontweight='bold')
axes[0,2].set_ylabel('Recall')
axes[0,2].set_ylim(0, 1)
for bar in bars3:
    height = bar.get_height()
    axes[0,2].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                  f'{height:.3f}', ha='center', va='bottom')

# AUC-ROC comparison
auc_scores = [comparison_df.loc[model, 'AUC-ROC'] for model in models_names]
bars4 = axes[1,0].bar(models_names, auc_scores, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
axes[1,0].set_title('Model AUC-ROC Comparison', fontsize=14, fontweight='bold')
axes[1,0].set_ylabel('AUC-ROC')
axes[1,0].set_ylim(0, 1)
for bar in bars4:
    height = bar.get_height()
    axes[1,0].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                  f'{height:.3f}', ha='center', va='bottom')

# Training History - Loss
for model_name in models_names:
    axes[1,1].plot(results[model_name]['history'].history['loss'], 
                  label=f'{model_name} Train', alpha=0.7)
axes[1,1].set_title('Training Loss Over Epochs', fontsize=14, fontweight='bold')
axes[1,1].set_ylabel('Loss')
axes[1,1].set_xlabel('Epoch')
axes[1,1].legend()

# Training History - Validation Loss
for model_name in models_names:
    axes[1,2].plot(results[model_name]['history'].history['val_loss'], 
                  label=f'{model_name} Val', linestyle='--', alpha=0.7)
axes[1,2].set_title('Validation Loss Over Epochs', fontsize=14, fontweight='bold')
axes[1,2].set_ylabel('Loss')
axes[1,2].set_xlabel('Epoch')
axes[1,2].legend()

plt.tight_layout()
plt.show()

# Cell 15: ROC Curves
plt.figure(figsize=(10, 8))
for model_name, metrics in results.items():
    fpr, tpr, _ = roc_curve(y_test, metrics['y_pred_proba'])
    auc_score = metrics['auc_roc']
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc_score:.4f})', linewidth=2.5)

plt.plot([0, 1], [0, 1], 'k--', alpha=0.5, label='Random Classifier')
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curve Comparison - All Models', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.show()




```
