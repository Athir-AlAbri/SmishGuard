## Install & Import Libraries

```python
# === Cell 1: Imports & Setup ===

import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.optim import AdamW
from transformers import get_linear_schedule_with_warmup
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
import os
from tqdm.auto import tqdm
import random

import matplotlib.pyplot as plt
import seaborn as sns

```

## Set Seeds

```python
# === Cell 2: Set Random Seeds ===

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(42)
```

## Load Augmented Dataset (for Token Length Analysis)

```python
# === Cell 3: Load augmented dataset for token analysis ===
df_aug = pd.read_csv("prep_out/augmented_full_dataset.csv")
print("Augmented dataset shape:", df_aug.shape)
print(df_aug.head(3))
print("Label distribution:", df_aug["label"].value_counts().to_dict())
```

## Analyze DistilBERT Token Length Distribution (Augmented Dataset)

```python
# === Cell 4: Token length distribution (WITH special tokens) ===

model_name = "distilbert-base-uncased"
tokenizer_len = AutoTokenizer.from_pretrained(model_name)

print("Computing token lengths...")

df_aug["token_len"] = df_aug["text"].astype(str).apply(
    lambda x: len(
        tokenizer_len(
            x,
            add_special_tokens=True,
            truncation=False
        )["input_ids"]
    )
)

print("\n=== TOKEN LENGTH SUMMARY ===")
print(df_aug["token_len"].describe())

for p in [50, 75, 90, 95, 99]:
    value = np.percentile(df_aug["token_len"], p)
    print(f"{p}th percentile: {value:.2f} tokens")

print("\nMax token length:", df_aug["token_len"].max())

plt.figure(figsize=(8, 5))
plt.hist(df_aug["token_len"], bins=40, edgecolor="black")
plt.title("Token Length Distribution (DistilBERT with Special Tokens)")
plt.xlabel("Number of tokens")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.5)
plt.xlim(0, df_aug["token_len"].max() + 5)
plt.show()

```

## Recommend Best max_length Based on Token Distribution

```python
# === Cell 5: Recommend best max_length ===

p95 = int(np.ceil(np.percentile(df_aug["token_len"], 95)))
p99 = int(np.ceil(np.percentile(df_aug["token_len"], 99)))
max_seen = int(df_aug["token_len"].max())

print("95th percentile:", p95)
print("99th percentile:", p99)
print("Max token length:", max_seen)

# Recommend based on 99th percentile rounded up to a convenient bucket
if p99 <= 70:
    MAX_LEN = 70
elif p99 <= 95:
    MAX_LEN = 95
elif p99 <= 128:
    MAX_LEN = 128
else:
    MAX_LEN = 256

print(f"\nRecommended MAX_LEN = {MAX_LEN}")
print(f"Rationale: Covers ~99% of token lengths including special tokens (p99={p99}),")
print("while minimizing padding and computational overhead for SMS data.")
```

## Inspect Truncation Effect + padding

```python
# === Cell 6: Inspect truncation effect on longest messages (shows special tokens) ===

def inspect_truncation(df, tokenizer, max_len=MAX_LEN, top_n=5, show_chars=300):
    """
    Shows how the longest messages get truncated at max_len.
    Prints both decoded versions:
      - WITH special tokens  ([CLS], [SEP], [PAD])
      - WITHOUT special tokens (clean text)
    """
    if "token_len" not in df.columns:
        raise ValueError("Column 'token_len' not found. Run the token length cell first.")

    longest = df.sort_values("token_len", ascending=False).head(top_n)

    for i, row in enumerate(longest.itertuples(index=False), start=1):
        text = str(row.text)
        true_len = int(getattr(row, "token_len"))

        print("\n" + "=" * 90)
        print(f"[{i}/{top_n}] Original token length: {true_len}")
        print(f"Original text (first {show_chars} chars):")
        print(text[:show_chars])

        # Encode EXACTLY like training (adds special tokens, pads, truncates)
        enc = tokenizer(
            text,
            add_special_tokens=True,
            truncation=True,
            padding="max_length",
            max_length=max_len,
            return_tensors="pt"
        )

        ids = enc["input_ids"][0].tolist()
        tokens = tokenizer.convert_ids_to_tokens(ids)

        print("\n--- AFTER TRUNCATION ---")
        print(f"MAX_LEN used: {max_len}")
        print(f"Encoded length (input_ids): {len(ids)}")

        # Show presence of special tokens
        print("\nFirst 25 tokens:")
        print(tokens[:25])
        print("\nLast 25 tokens:")
        print(tokens[-25:])

        # Decode WITH special tokens (so you can see [CLS], [SEP], [PAD])
        decoded_with_special = tokenizer.decode(ids, skip_special_tokens=False)
        print(f"\nDecoded WITH special tokens (first {show_chars} chars):")
        print(decoded_with_special[:show_chars])

        # Decode WITHOUT special tokens (clean readable text)
        decoded_clean = tokenizer.decode(ids, skip_special_tokens=True)
        print(f"\nDecoded WITHOUT special tokens (first {show_chars} chars):")
        print(decoded_clean[:show_chars])

        # How much is padding?
        pad_id = tokenizer.pad_token_id
        pad_count = sum(1 for t in ids if t == pad_id)
        print(f"\nPadding tokens: {pad_count} / {max_len}")

# Run the inspection on your augmented dataset
inspect_truncation(df_aug, tokenizer_len, max_len=MAX_LEN, top_n=5, show_chars=300)
```

```python
# For padding, just to proves that padding happens:
inspect_truncation(df_aug.sort_values("token_len").head(5), tokenizer_len, max_len=MAX_LEN, top_n=5)
```

## SMSDataset Class

```python
# === Cell 7: SMSDataset Class ===

class SMSDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=MAX_LEN):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = self.labels[idx]
        
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }
```

## Load Train/Val/Test CSVs

```python
# === Cell 8: Load Datasets ===

def load_datasets():
    train_df = pd.read_csv("prep_out/train.csv")
    val_df   = pd.read_csv("prep_out/val.csv")
    test_df  = pd.read_csv("prep_out/test.csv")
    
    print(f"Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")
    print("Train class distribution:", train_df['label'].value_counts().to_dict())
    return train_df, val_df, test_df
```

## Metrics Function

```python
# === Cell 9: Compute Metrics ===

def compute_metrics(predictions, labels):
    predictions = np.argmax(predictions, axis=1)
    
    accuracy = accuracy_score(labels, predictions)
    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, predictions, average='binary', zero_division=0
    )
    
    return {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1
    }
```

## Plot Helpers (Confusion Heatmap + Loss Curves)

```python
# === Cell 10: Plot helpers ===

def plot_loss_curves(history_df):
    plt.figure(figsize=(7, 5))
    plt.plot(history_df["Epoch"], history_df["Training Loss"], marker='o', label="Training Loss")
    plt.plot(history_df["Epoch"], history_df["Validation Loss"], marker='o', label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training vs Validation Loss")
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_confusion_matrix(cm):
    plt.figure(figsize=(6, 4))
    sns.heatmap(
        cm, annot=True, fmt="d", cmap="Blues",
        xticklabels=["Pred Ham (0)", "Pred Smish (1)"],
        yticklabels=["True Ham (0)", "True Smish (1)"],
        square=True, linewidths=1, linecolor="black"
    )
    plt.title("Confusion Matrix (Test Set)")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.tight_layout()
    plt.show()
```

## Training Function

```python
# === Cell 11: Training function ===
def train_model():
    print("=== SMISH DETECTION - MODEL TRAINING ===")
    print(f" Using MAX_LEN = {MAX_LEN}")
    
    train_df, val_df, test_df = load_datasets()
    
    model_name = "distilbert-base-uncased"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name,
        num_labels=2,
        id2label={0: "ham", 1: "smish"},
        label2id={"ham": 0, "smish": 1}
    )
    
    train_dataset = SMSDataset(train_df['text'].values, train_df['label'].values, tokenizer, max_length=MAX_LEN)
    val_dataset   = SMSDataset(val_df['text'].values,   val_df['label'].values,   tokenizer, max_length=MAX_LEN)
    test_dataset  = SMSDataset(test_df['text'].values,  test_df['label'].values,  tokenizer, max_length=MAX_LEN)
    
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader   = DataLoader(val_dataset, batch_size=32, shuffle=False)
    test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    optimizer = AdamW(model.parameters(), lr=2e-5, weight_decay=0.01)
    total_steps = len(train_loader) * 4
    scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=100, num_training_steps=total_steps)
    
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f" Using device: {device}")
    model.to(device)
    
    best_f1 = 0.0
    patience = 2
    no_improve_count = 0
    history = []
    
    for epoch in range(1, 5):
        print(f"\n=== Epoch {epoch}/4 ===")
        
        model.train()
        total_train_loss = 0.0
        for batch in tqdm(train_loader, desc=f"Training Epoch {epoch}"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels         = batch['labels'].to(device)
            
            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            
            loss.backward()
            optimizer.step()
            scheduler.step()
            optimizer.zero_grad()
            
            total_train_loss += loss.item()
        
        avg_train_loss = total_train_loss / len(train_loader)
        
        model.eval()
        total_val_loss = 0.0
        val_logits, val_true = [], []
        
        with torch.no_grad():
            for batch in tqdm(val_loader, desc=f"Validation Epoch {epoch}"):
                input_ids      = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels         = batch['labels'].to(device)
                
                outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
                loss = outputs.loss
                total_val_loss += loss.item()
                
                val_logits.extend(outputs.logits.cpu().numpy())
                val_true.extend(labels.cpu().numpy())
        
        avg_val_loss = total_val_loss / len(val_loader)
        val_metrics = compute_metrics(np.array(val_logits), np.array(val_true))
        
        print(f"Train Loss: {avg_train_loss:.6f} | Val Loss: {avg_val_loss:.6f} | Val F1: {val_metrics['f1']:.6f}")
        
        history.append({
            "Epoch": epoch,
            "Training Loss": round(avg_train_loss, 6),
            "Validation Loss": round(avg_val_loss, 6),
            "Accuracy": round(val_metrics["accuracy"], 6),
            "Precision": round(val_metrics["precision"], 6),
            "Recall": round(val_metrics["recall"], 6),
            "F1": round(val_metrics["f1"], 6),
        })
        
        if val_metrics["f1"] > best_f1:
            best_f1 = val_metrics["f1"]
            no_improve_count = 0
            save_dir = "smish_detection_model/final_model"
            os.makedirs(save_dir, exist_ok=True)
            model.save_pretrained(save_dir)
            tokenizer.save_pretrained(save_dir)
            print(f" Best model saved to {save_dir}")
        else:
            no_improve_count += 1
            if no_improve_count >= patience:
                print(" Early stopping triggered.")
                break
    
    history_df = pd.DataFrame(history)
    display(history_df)
    
    # TEST EVALUATION
    model.eval()
    test_logits, test_true = [], []
    with torch.no_grad():
        for batch in tqdm(test_loader, desc="Testing"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels         = batch['labels'].to(device)
            
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            test_logits.extend(outputs.logits.cpu().numpy())
            test_true.extend(labels.cpu().numpy())
    
    test_logits_arr = np.array(test_logits)
    test_true_arr   = np.array(test_true)
    
    test_metrics = compute_metrics(test_logits_arr, test_true_arr)
    print("\n Final test metrics:", test_metrics)
    
    # Confusion matrix + heatmap
    test_preds = np.argmax(test_logits_arr, axis=1)
    cm = confusion_matrix(test_true_arr, test_preds, labels=[0, 1])
    print("\nConfusion Matrix:\n", cm)
    plot_confusion_matrix(cm)
    
    return model, tokenizer, history_df

print(" train_model() defined.")
```

## Run Training

```python
# === Cell 12: Run training ===

model, tokenizer, history_df = train_model()

# Overfitting graph
plot_loss_curves(history_df)
```
